```python
import pandas as pd
import numpy as np
```

```python
def not_exist_searcher(comp:list[int], lack:list[int]) -> list[int]:
    a = comp.copy()
    b = lack.copy()
    not_exist = []
    for _ in range(len(b)):
        try:
            if a[0] == b[0]:
                a = np.delete(a, 0)
                b = np.delete(b, 0)
            if a[0] != b[0]:
                not_exist.append(a[0].item())
                a = np.delete(a, 0)
        except IndexError:
            pass
    return np.array(not_exist)
```

### product_id を int型にする

```python
df = pd.read_csv('df_user.csv', index_col='Unnamed: 0')
df['product_int'] = [int(data[:-2]) for data in df.product_id.values]
df.head(5)
```

### 目的のarrayの形状を全てNoneで生成（欠番がない状態で生成する）

```python
columns = np.sort(df.product_int.unique())
columns_length = np.arange((columns.max() + 1))

index = df.user_id.unique()
index_length = np.arange((index.max() + 1))

nan_frame = np.full((max(index)+1, max(columns)+1), None)
```

### 生成したarrayに対応する場所を上書きする

```python
for user_id, product_id, score in zip(df.user_id, df.product_int, df.score):
    nan_frame[user_id, product_id] = score
```

### arrayからDataFrameを作成

```python
data_frame = pd.DataFrame(nan_frame, index=index_length, columns=columns_length)
```

### 欠落したcolumnやindexを削除

```python
# 欠落カラムの選別
not_columns = not_exist_searcher(columns_length, columns)

# 欠落インデックスの選別
not_indices = not_exist_searcher(index_length, index)
```

## 存在しないカラムとインデックスを削除

```python
data_frame = data_frame.drop(not_columns, axis=1)
data_frame = data_frame.drop(not_indices, axis=0)
```
